# Oversampling 

```python
# Filter out by Exited and store each class in a different variable
major_df = data.filter(col("Exited") == 0)
minor_df = data.filter(col("Exited") == 1)

# Calculate the ratio
ratio = int(major_df.count()/minor_df.count())

# Print
print("ratio: {}".format(ratio))
```

```python
# Store range (0,3)
a = range(ratio)

# duplicate the minority rows
oversampled_df = minor_df.withColumn("dummy", explode(array([lit(x) for x in a]))).drop('dummy')

# combine both oversampled minority rows and previous majority rows 
combined_df = major_df.unionAll(oversampled_df)

# Show df
combined_df.show()
```

```python
# Select column and store in a new variable
df_3 = combined_df.select(combined_df["Exited"]).orderBy('Exited', ascending=True)

# Plot Histogram
fig = px.histogram(df_3.toPandas(),
                   x="Exited", 
                   color="Exited",
                   color_discrete_sequence=["#357EC7", "#C35817"])
# Display Figure
fig.show()
```

```python
# Store distinct Gender distinct values
distinct_values = combined_df.select("Gender")\
                    .distinct()\
                    .rdd\
                    .flatMap(lambda x: x).collect()

# For each of the gathered values create a new column 
for distinct_value in distinct_values:
    function = udf(lambda item: 
                   1 if item == distinct_value else 0, 
                   IntegerType())
    new_column_name = "Gender"+'_'+ distinct_value
    df_final = combined_df.withColumn(new_column_name, function(col("Gender")))

# Print
df_final.show()
```

# ANN Binary Classifier

Data related to Germany is going to be used for predictions because is the country that has the most balanced data as can be seen in the last stacked bar chart above. Moreover, this technique can be applied to the other countries however with no guarantee that the model will perform well due to the unbalanced data in the Exited column.

```python
# List of Predictors
Predictors = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 
            'HasCrCard', 'IsActiveMember', 'EstimatedSalary', 'Gender_male']

# Store the list into the X variable 
X = df_germany[Predictors].toPandas()
```

```python
# Separate Target Variable and Predictor Variables

# Target
TargetVariable = ['Exited']

# Predictors
Predictors = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 
            'HasCrCard', 'IsActiveMember', 'EstimatedSalary', 'Gender_male']
 
# Transform Predictors into 2D array
X = df_germany[Predictors].toPandas()
X = X.values

# Transform Target variable into 2D array
y = df_germany[TargetVariable].toPandas()
y = y.values
 
# StandardScaler
PredictorScaler = StandardScaler()
 
# Storing the fit object for later reference
PredictorScalerFit = PredictorScaler.fit(X)
 
# Apply StandardScaler on the Predictors
X = PredictorScalerFit.transform(X)
 
# Train and test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
 
# Print shape of the training and test set
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

# Simple ANN Classifier model

```python
# Create Sequential
classifier = Sequential()

# Define the Input hidden layer and first layer and activation function RELU
classifier.add(Dense(units=10, input_dim=9, kernel_initializer='uniform', activation='relu'))
 
#Define the second hidden layer
classifier.add(Dense(units=6, kernel_initializer='uniform', activation='relu'))
 
# Define the Output layer and sigmoid means sigmoid as activation function
classifier.add(Dense(units=1, kernel_initializer='uniform', activation='sigmoid'))
 
# Optimizer Adam and Metric chosen is accurary 
classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
 
# Fit the Neural Network on the training data
ANN_Model=classifier.fit(X_train,y_train, batch_size=10 , epochs=10, verbose=1)

```

# GridSearchCV & Best Paramaters

```python
# Define function to find the best parameters
def ANN_Best_Params(X_train, y_train):
    
    # Create listed with potential numbers for Trial, batch size and epoch list
    
    # Number of trials
    TrialNumber=0

    # Batch size
    batch_size_list=[5, 10, 15, 20]

    # Training size
    epoch_list=[5, 10, 50 ,100]
    
    # Import pandas
    import pandas as pd

    # Create a dataframe 
    SearchResultsData=pd.DataFrame(columns=['TrialNumber', 'Parameters', 'Accuracy'])
    
    # Loop over the lists
    for batch_size_trial in batch_size_list:
        for epochs_trial in epoch_list:
            TrialNumber+=1
            
            # Create the ANN Classifier based on the values
            classifier = Sequential()

            # First Layer
            classifier.add(Dense(units=10, input_dim=9, kernel_initializer='uniform', activation='relu'))

            # Second Layer
            classifier.add(Dense(units=6, kernel_initializer='uniform', activation='relu'))

            # Output Layer
            classifier.add(Dense(units=1, kernel_initializer='uniform', activation='sigmoid'))

            # Compile model
            classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
 
            
            # Train the ANN model on training data
            ANN_Model=classifier.fit(X_train,y_train, batch_size=batch_size_trial , epochs=epochs_trial, verbose=0)
            
            # Fetch accurary of the model
            Accuracy = ANN_Model.history['accuracy'][-1]
            
            # Print results of the current iterations
            print(TrialNumber, 'Parameters:','batch_size:', batch_size_trial,'-', 'epochs:',epochs_trial, 'Accuracy:', Accuracy)
            
            # Append
            SearchResultsData=SearchResultsData.append(pd.DataFrame(data=[[TrialNumber,
                            'batch_size'+str(batch_size_trial)+'-'+'epoch'+str(epochs_trial), Accuracy]],
     
                                                                    columns=['TrialNumber', 'Parameters', 'Accuracy'] ))
    return(SearchResultsData)
 
###############################################
 
# Call Function using the training and test set
ResultsData=ANN_Best_Params(X_train, y_train)
```

```python
# Train the models with the best parameters found
classifier.fit(X_train,y_train, batch_size=20 , epochs=100, verbose=1)
```

```python
Predictions=classifier.predict(X_test)
 
# Scale the test data back to original scale
Test_Data=PredictorScalerFit.inverse_transform(X_test)
 
# Generating a data frame for analyzing the test data
TestingData=pd.DataFrame(data=Test_Data, columns=Predictors)
TestingData['Exited']=y_test
TestingData['Predictions']=Predictions

# Defining the probability threshold
def probThreshold(inpProb):
    if inpProb > 0.5:
        return(1)
    else:
        return(0)
 
# Generating predictions on the testing data by applying probability threshold
TestingData['Predictions']=TestingData['Predictions'].apply(probThreshold)
print(TestingData.head())
 
###############################################
print('\n######### Testing Accuracy Results #########')
print(metrics.classification_report(TestingData['Exited'], TestingData['Predictions']))
print(metrics.confusion_matrix(TestingData['Exited'], TestingData['Predictions']))
```

# Machine Learning Pipeline


```python
# Create a Pipeline for Germany

def classifier_pipeline():

  # Start SparkSession
  spark = SparkSession.builder.appName('bdsp').getOrCreate()

  # Read of the data using read.csv 
  data = spark.read.csv("Churn_Modelling.csv",header=True)


  # Data transformation
  data = data.withColumn("EstimatedSalary", 
                         data["EstimatedSalary"].cast(FloatType()))
  data = data.withColumn("IsActiveMember", 
                         data["IsActiveMember"].cast(IntegerType()))
  data = data.withColumn("HasCrCard", 
                         data["HasCrCard"].cast(IntegerType()))
  data = data.withColumn("NumOfProducts", 
                         data["NumOfProducts"].cast(IntegerType()))
  data = data.withColumn("Balance", 
                         data["Balance"].cast(FloatType()))
  data = data.withColumn("Tenure", 
                         data["Tenure"].cast(IntegerType()))
  data = data.withColumn("Age", 
                         data["Age"].cast(IntegerType()))
  data = data.withColumn("CreditScore", 
                         data["CreditScore"].cast(IntegerType()))
  data = data.withColumn("CustomerId", 
                         data["CustomerId"].cast(IntegerType()))
  data = data.withColumn("RowNumber", 
                         data["RowNumber"].cast(IntegerType()))
  data = data.withColumn("Exited", 
                         data["Exited"].cast(IntegerType()))

  # Filter out by Exited and store each class in a different variable
  major_df = data.filter(col("Exited") == 0)
  minor_df = data.filter(col("Exited") == 1)

  # Calculate the ratio
  ratio = int(major_df.count()/minor_df.count())

  # Store range (0,3)
  a = range(ratio)

  # duplicate the minority rows
  oversampled_df = minor_df.withColumn("dummy", 
                                       explode(array([lit(x) for x in a])))\
                                       .drop('dummy')

  # combine both oversampled minority rows and previous majority rows 
  combined_df = major_df.unionAll(oversampled_df)

  # Store distinct Gender distinct values
  distinct_values = combined_df.select("Gender")\
                      .distinct()\
                      .rdd\
                      .flatMap(lambda x: x).collect()

  # For each of the gathered values create a new column 
  for distinct_value in distinct_values:
      function = udf(lambda item: 
                    1 if item == distinct_value else 0, 
                    IntegerType())
      new_column_name = "Gender"+'_'+ distinct_value
      df_final = combined_df.withColumn(new_column_name, 
                                        function(col("Gender")))

  # Limit dataset for only France
  df_country = df_final.filter(df_final.Geography == "Germany")

  # Drop columns
  cols = ("RowNumber", "CustomerId", "Surname", "Geography", "Gender")
  df_country.drop(*cols)

  # List of Predictors
  Predictors = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 
            'HasCrCard', 'IsActiveMember', 'EstimatedSalary', 'Gender_male']

  # Store the list into the X variable 
  X = df_country[Predictors].toPandas()

  # Separate Target Variable and Predictor Variables

  # Target
  TargetVariable = ['Exited']

  # Predictors
  Predictors = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 
              'HasCrCard', 'IsActiveMember', 'EstimatedSalary', 'Gender_male']
  
  # Transform Predictors into 2D array
  X = df_country[Predictors].toPandas()
  X = X.values

  # Transform Target variable into 2D array
  y = df_country[TargetVariable].toPandas()
  y = y.values
  
  # StandardScaler
  PredictorScaler = StandardScaler()
  
  # Storing the fit object for later reference
  PredictorScalerFit = PredictorScaler.fit(X)
  
  # Apply StandardScaler on the Predictors
  X = PredictorScalerFit.transform(X)
  
  # Train and test split
  X_train, X_test, y_train, y_test = train_test_split(X, 
                                                      y, 
                                                      test_size=0.3, 
                                                      random_state=42)

  # Load the model
  model = load_model("germany_model.h5")

  # Predictions
  model.predict(X_test)

  # Scale the test data back to original scale
  Test_Data=PredictorScalerFit.inverse_transform(X_test)
  
  # Generating a data frame for analyzing the test data
  TestingData=pd.DataFrame(data=Test_Data, columns=Predictors)
  TestingData['Exited']=y_test
  TestingData['Predictions']=Predictions

  # Defining the probability threshold
  def probThreshold(inpProb):
      if inpProb > 0.5:
          return(1)
      else:
          return(0)
  
  # Generating predictions on the testing data by applying probability threshold
  TestingData['Predictions']= TestingData['Predictions'].apply(probThreshold)
  print(TestingData.head())
  
  # Print classification report and confusion matrix
  print('\n--------- Testing Accuracy Results ----------')
  print(metrics.classification_report(TestingData['Exited'], 
                                      TestingData['Predictions']))
  print(metrics.confusion_matrix(TestingData['Exited'], 
  TestingData['Predictions']))
```
